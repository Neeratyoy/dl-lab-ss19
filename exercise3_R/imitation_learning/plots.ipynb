```python
import tensorflow as tf
def read_logs(file):
    logs = tf.train.summary_iterator(file)
    plot_data = {'train_loss_1': [], 'valid_loss_1': [],
                 'train_accuracy_1': [], 'valid_accuracy_1': []}
    for log in logs:
        values = log.summary.value
        for val in values:
            plot_data[val.tag].append(val.simple_value)
    return plot_data
```

```python
#################################
# Path to tensorflow event file #
#################################
log_file_name = '/media/neeratyoy/Mars/Freiburg/SummerSemester19/DL_Lab/dl-lab-ss19/exercise3_R/reinforcement_learning/tensorboard/cartpole/with_exploration/events.out.tfevents.1559761324.neeratyoy'
def read_logs(file):
    logs = tf.train.summary_iterator(file)
    plot_data = {'a_0_1': [], 'a_1_1': [],
                 'episode_reward_1': [], 'eval_reward_1': []}
    for log in logs:
        values = log.summary.value
        for val in values:
            plot_data[val.tag].append(val.simple_value)
    return plot_data
pd = read_logs(log_file_name) 
```
